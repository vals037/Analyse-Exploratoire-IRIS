###### - Analyse de la corrélation entre les caractéristiques

```python
# Convertir la colonne 'Species' en valeurs numériques
#species_mapping = {'Iris-setosa': 0, 'Iris-versicolor': 1, 'Iris-virginica': 2}
#df['Species'] = df['Species'].map(species_mapping)
```

```python
# Calculer la matrice de corrélation
correlation_matrix = df[['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm', ]].corr()
```

```python
# Visualiser la matrice de corrélation avec une heatmap
plt.figure(figsize=(10, 6))
plt.imshow(correlation_matrix, cmap='coolwarm', interpolation='nearest')
plt.colorbar(label='Correlation')
plt.title('Matrice de corrélation entre les caractéristiques')
plt.xticks(range(len(correlation_matrix)), correlation_matrix.columns, rotation=45)
plt.yticks(range(len(correlation_matrix)), correlation_matrix.columns)
plt.show()
```

###### Les valeurs de corrélation varient de -0.4 à 1. Une valeur de 1 indique une corrélation positive parfaite, -1 indique une corrélation négative parfaite, et 0 indique aucune corrélation linéaire. Plus la valeur est proche de 1 ou -1, plus la relation linéaire entre les variables est forte.

###### -  Comparaison des caractéristiques entre les espèces¶

```python
import seaborn as sns
# Comparaison des caractéristiques entre les espèces avec des box plots
plt.figure(figsize=(10, 6))
sns.boxplot(x='Species', y='SepalLengthCm', data=df)
plt.title('Comparaison de la longueur du sépale entre les espèces')
plt.show() 
```

###### Quartiles et médiane :

La boîte du boxplot représente le premier quartile (Q1) au troisième quartile (Q3), où 50% des données se situent. La ligne à l'intérieur de la boîte représente la médiane (Q2), qui est la valeur au milieu de l'échantillon.
La longueur de la boîte (interquartile range, IQR) est une mesure de dispersion qui indique la dispersion des valeurs autour de la médiane.

###### Moustaches :

Les moustaches du boxplot représentent la plage des données en dehors du premier et du troisième quartile. Elles s'étendent généralement jusqu'à 1,5 fois l'IQR de la boîte.
Les points situés au-delà des moustaches sont généralement considérés comme des valeurs aberrantes mais tout reste dans l'objectif à atteindre.

###### Distribution des données :

La forme et la longueur de la boîte, ainsi que la distance entre les quartiles, donnent une idée de la distribution des données. Par exemple, une boîte plus longue indique une plus grande dispersion des données, tandis qu'une boîte plus courte indique une distribution plus concentrée.

###### Symétrie et skewness :

La symétrie de la boîte peut fournir des informations sur la distribution des données. Si la médiane est au centre de la boîte et les moustaches sont de longueurs égales des deux côtés, la distribution est symétrique. Sinon, la distribution peut être asymétrique (skewed).
Une asymétrie vers la gauche signifie que la queue de la distribution est plus longue à gauche, tandis qu'une asymétrie vers la droite signifie que la queue de la distribution est plus longue à droite.

###### Valeurs aberrantes :

Les points situés en dehors des moustaches sont considérés comme des valeurs aberrantes potentielles. Ces valeurs peuvent être des observations rares ou des erreurs de mesure.

```python
# Comparaison des caractéristiques entre les espèces avec des violin plots
plt.figure(figsize=(10, 6))
sns.violinplot(x='Species', y='SepalWidthCm', data=df)
plt.title('Comparaison de la largeur du sépale entre les espèces')
plt.show()
```

###### Forme des violons :

Chaque "violon" dans le graphique représente la distribution des données pour une espèce spécifique.
La largeur du violon représente la densité des données à différents niveaux de la variable. Un violon plus large indique une densité plus élevée des données à cette valeur de la variable.

###### Quartiles et médiane :

À l'intérieur de chaque violon, il y a un petit trait horizontal qui représente la médiane de la distribution.
La partie la plus large du violon représente la plage des valeurs où la plupart des données se trouvent, tandis que les parties plus étroites du violon représentent les queues de distribution où les données sont moins fréquentes.

###### Comparaison des distributions :

Un avantage du violinplot par rapport au boxplot est qu'il permet de comparer visuellement les distributions entre différentes catégories ou groupes.
Les violons peuvent être côte à côte pour différentes catégories, ce qui facilite la comparaison des distributions entre ces catégories.

###### Symétrie et skewness :

Comme pour un boxplot, la symétrie et la skewness (asymétrie) de la distribution peuvent être observées dans un violinplot.
Si la distribution est symétrique, le violon aura une forme symétrique. Si la distribution est asymétrique, le violon sera incliné dans la direction de l'asymétrie.

```python
# Comparaison des caractéristiques entre les espèces avec des swarm plots
plt.figure(figsize=(12, 8))
sns.swarmplot(x='Species', y='PetalLengthCm', data=df)
plt.title('Comparaison de la longueur du pétale entre les espèces')
plt.show() 
```

###### Répartition des données :

Chaque point dans le graphique représente une observation individuelle dans l'ensemble de données.
Les points sont répartis le long de l'axe de la variable continue, en fonction de leurs valeurs respectives.

###### Comparaison entre les catégories :

Les points sont regroupés horizontalement en fonction des différentes catégories ou groupes de la variable catégorielle.
Les points ne se chevauchent pas, ce qui permet de visualiser clairement la répartition des valeurs pour chaque catégorie.

###### Densité de données :

Comme les points ne se chevauchent pas, un swarmplot permet de visualiser la densité des données pour chaque catégorie.
Une plus grande densité de points dans une zone spécifique indique une concentration plus élevée de valeurs dans cette zone.

###### Identification des valeurs aberrantes :

Les points situés à l'extérieur du regroupement principal peuvent représenter des valeurs aberrantes ou des observations inhabituelles.
Le swarmplot permet de repérer facilement ces valeurs aberrantes en les isolant visuellement des autres observations.
